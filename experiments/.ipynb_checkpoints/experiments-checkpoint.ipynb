```python
import sys
sys.path.insert(0,'..')

import numpy as np
from markowitz import test_Markowitz, Markowitz
import matplotlib.pyplot as plt

P = np.load('../data/s&p500-2018-None.npy') # daily data from 2018-2020 on stock price
N, T = P.shape
T = T - 1
R = np.zeros((N,T))
R = (np.log(P[:,1:]) - np.log(P[:,:-1]))
n = 50
t = 240
sigma = 0.02
R = R[np.random.choice(R.shape[0], n, replace=False), :]

Rp_m, Rp_v, Rp_e = test_Markowitz(R, t, sigma)

plt.figure(figsize=(15,5))
plt.plot(Rp_m,color='r')
plt.plot(Rp_v,color='green')
plt.plot(Rp_e)
plt.show()
print("MAXSER: ",np.mean(Rp_m), np.sqrt(np.var(Rp_m)), np.mean(Rp_m)/np.sqrt(np.var(Rp_m)))
print("Vanilla",np.mean(Rp_v), np.sqrt(np.var(Rp_v)), np.mean(Rp_v)/np.sqrt(np.var(Rp_v)))
print("Equally weigthed",np.mean(Rp_e), np.sqrt(np.var(Rp_e)), np.mean(Rp_e)/np.sqrt(np.var(Rp_e)))
```
